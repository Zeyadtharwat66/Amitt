```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
dataset = pd.read_csv(r'C:\Users\Lenovo\Desktop\AMIT\adel\datadata.txt')
dataset
```

```python
df = dataset.copy()
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x
```

```python
y
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge 
```

```python
l=50
lasso = Lasso(alpha=l) 
lasso.fit(x,y)
```

```python
lasso_coeff = lasso.coef_
lasso_coeff 
```

```python
lasso_inter=lasso.intercept_
lasso_inter
```

```python
x_test = [1,2,3,4]
y_pred=lasso.predict(pd.DataFrame(x_test))
y_pred
```

```python
df
```

```python
plt.figure(figsize=(12,6) , dpi = 90)
plt.plot(x_test,y_pred , c ='r' , linewidth = 2)
plt.scatter(x,y)
plt.ylim(ymin = 0 , ymax = 9)
plt.xlim(xmin = 0 , xmax = 7)
plt.text(x_test[-1],y_pred[-1] , ' y = '   +    str('%.2f' %lasso_coeff)  + ' * X +'  +  str('%.2f' %lasso_inter)    
                                                    +  '  For \u03BB or  \u03b1 = ' +    str(l)  ,   fontsize = 12)
plt.title('Lasso Regression')
plt.show()
```

```python
l=2
ridge = Ridge(alpha=l) 
ridge.fit(x,y)
```

```python
ridge_coeff = ridge.coef_
ridge_coeff
```

```python
ridge_intercept = ridge.intercept_
ridge_intercept
```

```python
x_test = [1,2,3,4]
y_pred=ridge.predict(pd.DataFrame(x_test))
y_pred
```

```python
plt.figure(figsize=(12,6) , dpi = 90)
plt.plot(x_test,y_pred , c ='r' , linewidth = 2)
plt.scatter(x,y)
plt.ylim(ymin = 0 , ymax = 9)
plt.xlim(xmin = 0 , xmax = 7)
plt.text(x_test[-1],y_pred[-1] , ' y = '   +    str('%.2f' %ridge_coeff)  + ' * X +'  +  str('%.2f' %ridge_intercept)    
                                                    +  '  For \u03BB or  \u03b1 = ' +    str(l)  ,   fontsize = 12)
plt.title('ridge Regression')
plt.show()
```
