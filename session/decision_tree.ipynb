```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

dataset_path=(r'C:\D\Amitt\session\Position_Salaries.csv')
datasets=pd.read_csv(dataset_path)

datasets.head()
```

```python
x=datasets.iloc[:,1:2].values
y=datasets.iloc[:,2].values
```

```python
regressor=DecisionTreeRegressor(random_state=0)
regressor.fit(x,y)
```

```python
value_to_predict=6.5
y_pred=regressor.predict([[value_to_predict]])
y_pred[0]
```

```python
x_grid=np.arange(min(x),max(x),0.01)
x_grid=x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,color='red',label='Actual data')
plt.plot(x_grid,regressor.predict(x_grid),color='blue',label='Regression Curve')
plt.title('decision tree regression results')
plt.xlabel('position level')
plt.ylabel('salary')
plt.legend()
plt.grid()
plt.show()
```
