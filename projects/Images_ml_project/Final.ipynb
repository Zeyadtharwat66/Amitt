```python
import os
from scipy.io import loadmat
directory = r'C:\D\Amitt\projects\Images_ml_project\AFLW2000'
pose_para=[]
for filename in os.listdir(directory):
    if filename.endswith('.mat'):
        filepath = os.path.join(directory, filename)
        data = loadmat(filepath)
        if 'Pose_Para' in data:
            pose_para.append(data['Pose_Para'])
        else:
            print("Variable 'Pose_Para' not found in this file.")
```

```python
yaw=[]
pitch=[]
roll=[]
for elements in pose_para :
    yaw.append(elements[0][0])
    pitch.append(elements[0][1])
    roll.append(elements[0][2])
```

```python
import os
import cv2
directory = r'C:\D\Amitt\projects\Images_ml_project\AFLW2000'
images = []
for filename in os.listdir(directory):
    if filename.endswith('.jpg'):
        filepath = os.path.join(directory, filename)
        img = cv2.imread(filepath)
        if img is not None:
            images.append((filepath, img))
        else:
            print(f"Failed to load image: {filepath}")
```

```python
yaw
```

```python
import cv2
import mediapipe as mp
mp_face_mesh = mp.solutions.face_mesh
face_mesh_model = mp_face_mesh.FaceMesh(
    min_detection_confidence=0.2,
    min_tracking_confidence=0.2
)
mp_drawing = mp.solutions.drawing_utils
count=0
landmark_indices = [1, 152, 33, 263, 61, 291, 10, 234, 454]
landmarks_dataX = {}
landmarks_dataY = {}
for indx,(file_path, image) in enumerate (images):
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = face_mesh_model.process(rgb_image)
    image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2BGR)
    if results.multi_face_landmarks:
        count+=1
        image_landmarksX = []
        image_landmarksY = []
        for face_landmarks in results.multi_face_landmarks:
            mp_drawing.draw_landmarks(
                image,
                face_landmarks,
                connections=None,
                landmark_drawing_spec=mp_drawing.DrawingSpec(color=(255, 255, 255), thickness=-3 , circle_radius=3)
            )
            for idx in landmark_indices:
                l = face_landmarks.landmark[idx]
                image_landmarksX.append(l.x)
            for idx in landmark_indices:
                l = face_landmarks.landmark[idx]
                image_landmarksY.append(l.y)
        landmarks_dataX[file_path] = image_landmarksX
        landmarks_dataY[file_path] = image_landmarksY
    else :
        landmarks_dataX[file_path] = "NULL"
        landmarks_dataY[file_path] = "NULL"
    cv2.imwrite(file_path, image)
count
```

```python
print(landmarks_dataX)
```

```python
import pandas as pd
f=pd.DataFrame(landmarks_dataX)
transposed_dfX=f.T
transposed_dfX.columns = ['x0', 'x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8']
x=pd.DataFrame(landmarks_dataY)
transposed_dfY=x.T
transposed_dfY.columns = ['y0', 'y1', 'y2', 'y3', 'y4', 'y5', 'y6', 'y7', 'y8']
data=pd.DataFrame({'yaw':yaw,'pitch':pitch,'roll':roll},)
transposed_dfX.reset_index(drop=True, inplace=True)
transposed_dfY.reset_index(drop=True, inplace=True)
data.reset_index(drop=True, inplace=True)
final_data =pd.concat([transposed_dfX, transposed_dfY, data], axis=1)
final_data.head(20)
```

```python
df_cleaned = final_data[final_data != 'NULL']
df_cleaned.dropna(inplace=True)
```

```python
df_cleaned
```

```python
X=df_cleaned.drop(['yaw', 'pitch', 'roll'],axis=1)
y_model1_yaw=df_cleaned['yaw']
y_model1_pitch=df_cleaned['pitch']
y_model1_roll=df_cleaned['roll']
from sklearn.model_selection import train_test_split
x_train_yaw, x_test_yaw, y_train_yaw, y_test_yaw = train_test_split(X, y_model1_yaw, test_size=0.3, random_state=0)
x_train_pitch, x_test_pitch, y_train_pitch, y_test_pitch = train_test_split(X, y_model1_pitch, test_size=0.3, random_state=0)
x_train_roll, x_test_roll, y_train_roll, y_test_roll = train_test_split(X, y_model1_roll, test_size=0.3, random_state=0)
```

Yaw

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model = LinearRegression()
model.fit(x_train_yaw, y_train_yaw)
y_pred_yaw = model.predict(x_test_yaw)
mse_yaw = mean_squared_error(y_test_yaw, y_pred_yaw) 
rmse_yaw = np.sqrt(mse_yaw) 
r2_yaw = r2_score(y_test_yaw, y_pred_yaw)
mse_yaw
```

Pitch

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model = LinearRegression()
model.fit(x_train_pitch, y_train_pitch)
y_pred_pitch = model.predict(x_test_pitch)
mse_pitch = mean_squared_error(y_test_pitch, y_pred_pitch)
rmse_pitch = np.sqrt(mse_pitch)
r2_pitch = r2_score(y_test_pitch, y_pred_pitch)
mse_pitch
```
